```python
#implement a print_words(filename) function that counts how often each word appears in the text and prints:
#word1 count1
#word2 count2
#Print the above list in order sorted by word


def print_words(filename):
    lst = []
    string = ''
    dictionary = {}
    with open (filename, 'r') as x:
        for i in x:
            lst.append(i)
        for i in lst:
            string = string + i
        lst_new = string.split(' ')
        lst_new.sort()
        for i in lst_new:
            a = lst_new.count(i)
            dictionary[i] = a
        return dictionary
        
        
print_words('alice.txt')
```
